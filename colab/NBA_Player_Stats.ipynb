```python
from pandas.core.frame import DataFrame
import pandas as pd
import math

def normal_round(n, decimals=0):
    expoN = n * 10 ** decimals
    if abs(expoN) - abs(math.floor(expoN)) < 0.5:
        return math.floor(expoN) / 10 ** decimals
    return math.ceil(expoN) / 10 ** decimals

def calc_percent(value):
  if math.isnan(value):
    return 0

  return normal_round(value, 2) * 100

def has_double(stat_count):
  return stat_count >= 10

def calc_double_points(dbl_count_max, score, points, rebounds, assists, steals, blocks):
  dbl_count = 0

  if has_double(points):
    dbl_count = dbl_count + 1

  if has_double(rebounds):
    dbl_count = dbl_count + 1

  if has_double(assists):
    dbl_count = dbl_count + 1

  if has_double(steals):
    dbl_count = dbl_count + 1

  if has_double(blocks):
    dbl_count = dbl_count + 1

  if dbl_count >= dbl_count_max:
    return score
  else:
    return 0

players_df = pd.read_csv(f"https://raw.githubusercontent.com/kbw612/Fantasy/main/NBA/2023-24/NBA-Players.csv")

players_df["STOCKS"] = 0
players_df["DK_FPTS"] = 0.0

for row in players_df.index:
  players_df.at[row, "STOCKS"] = players_df.at[row, "STL"] + players_df.at[row, "BLK"]
  three_pointers_made_points = players_df.at[row, "3PM"] * .5
  rebound_points = players_df.at[row, "REB"] * 1.25
  assist_points = players_df.at[row, "AST"] * 1.5
  steal_points = players_df.at[row, "STL"] * 2
  block_points = players_df.at[row, "BLK"] * 2
  turnover_points = players_df.at[row, "TOV"] * -.5
  double_double_points = calc_double_points(2, 1.5, players_df.at[row, "PTS"], players_df.at[row, "REB"], players_df.at[row, "AST"], players_df.at[row, "STL"], players_df.at[row, "BLK"])
  triple_double_points = calc_double_points(3, 3, players_df.at[row, "PTS"], players_df.at[row, "REB"], players_df.at[row, "AST"], players_df.at[row, "STL"], players_df.at[row, "BLK"])
  players_df.at[row, "DK_FPTS"] = normal_round(players_df.at[row, "PTS"] + three_pointers_made_points + rebound_points + assist_points + steal_points + block_points + turnover_points + double_double_points + triple_double_points, 1)
  players_df.at[row, "PTS_%"] = calc_percent(players_df.at[row, "PTS"] / players_df.at[row, "DK_FPTS"])
  players_df.at[row, "3PM_PTS"] = three_pointers_made_points
  players_df.at[row, "3PM_PTS_%"] = calc_percent(three_pointers_made_points / players_df.at[row, "DK_FPTS"])
  players_df.at[row, "RB_PTS"] = rebound_points
  players_df.at[row, "RB_PTS_%"] = calc_percent(rebound_points / players_df.at[row, "DK_FPTS"])
  players_df.at[row, "AST_PTS"] = assist_points
  players_df.at[row, "AST_PTS_%"] = calc_percent(assist_points / players_df.at[row, "DK_FPTS"])
  players_df.at[row, "STL_PTS"] = steal_points
  players_df.at[row, "STL_PTS_%"] = calc_percent(steal_points / players_df.at[row, "DK_FPTS"])
  players_df.at[row, "BLK_PTS"] = block_points
  players_df.at[row, "BLK_PTS_%"] = calc_percent(block_points / players_df.at[row, "DK_FPTS"])
  players_df.at[row, "TOV_PTS"] = turnover_points
  players_df.at[row, "TOV_PTS_%"] = calc_percent(turnover_points / players_df.at[row, "DK_FPTS"])
  players_df.at[row, "DBL_DBL_PTS"] = double_double_points
  players_df.at[row, "DBL_DBL_PTS_%"] = calc_percent(double_double_points / players_df.at[row, "DK_FPTS"])
  players_df.at[row, "TRP_DBL_PTS"] = triple_double_points
  players_df.at[row, "TRP_DBL_PTS_%"] = calc_percent(triple_double_points / players_df.at[row, "DK_FPTS"])

# 1 3PT .5 RB 1.25 assist 1.5 STL 2  BLK 2  TO -.5  1.5 dbl dbl 3 TPL DBL

with pd.option_context("display.max_rows", 1000):
  print(players_df.to_string(index=False))
  print("")

# PLAYER TEAM    MATCH UP  GAME DATE W/L  MIN  PTS  FGM  FGA  FG%  3PM  3PA  3P%  FTM  FTA  FT%  OREB  DREB  REB  AST  STL  BLK  TOV  PF  +/-   FP  STOCKS  DK_FPTS
#Nic Claxton  BKN BKN vs. MIN 01/25/2024   L   39   16    8   10   80    0    0    -    0    0    -     2     9   11    4    2    2    0   2    2 47.2       4    45.25

```
