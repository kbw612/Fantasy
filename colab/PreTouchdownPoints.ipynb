```python
import pandas as pd

class ReceivingStats:
  def __init__(self, receptions, yards, touchdowns):
      self.receptions = receptions
      self.yards = yards
      self.touchdowns = touchdowns

class RushingStats:
  def __init__(self, yards, touchdowns):
      self.yards = yards
      self.touchdowns = touchdowns

class GameStats:
  def __init__(self, receivingStats, rushingStats):
      self.receivingStats = receivingStats
      self.rushingStats = rushingStats

def calculateNonTouchdownPoints(rushingYards, receptions, receivingYards):
  return (rushingYards *.1) + (receptions * 1) + (receivingYards *.1)

def calculateNonTouchdownPoints2(fantasyPoints, receivingTouchtowns):
  return fantasyPoints - (receivingTouchtowns * 6)

current_week = 6

positions = ['WR']

#"Rank","Player","REC","TGT","YDS","Y/R","LG","20+","TD","ATT","YDS","TD","FL","G","FPTS","FPTS/G","ROST"

for position in positions:
  stats_df = pd.read_csv(f'https://raw.githubusercontent.com/kbw612/Fantasy/main/NFL/2023/stats/FantasyPros_Fantasy_Football_Statistics_{position}_week{current_week}.csv')

  stats_df['preTDPoints'] = calculateNonTouchdownPoints(stats_df['RUSHYDS'], stats_df['REC'], stats_df['RECYDS'])
  stats_df['preTDPoints2'] = calculateNonTouchdownPoints2(stats_df['FPTS'], stats_df['TD'])

  with pd.option_context('display.max_rows', 1000):
    print(stats_df.to_string(index=False))
```
